```python
def get_name(name_node):
  text = name_node.get_text()
  name = text[:text.find("(")].strip()
  return name

def get_country(name_node):
  country_text = name_node.get_text()
  country_item = country_text[country_text.find("(") + 1 : country_text.find(")")]
  return country_item
```
