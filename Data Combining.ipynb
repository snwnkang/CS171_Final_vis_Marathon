```python
london = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/combined_data/London_Marathon2014-2023.csv")
chicago = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/combined_data/chicago_marathon_2014-2023.csv")
newyork = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/combined_data/ny_marathon_2014-2023.csv")
```

```python
london = london.drop(columns=['Unnamed: 0'])
chicago = chicago.drop(columns=['Unnamed: 0'])
newyork = newyork.drop(columns=['Unnamed: 0'])
```

```python
london['Marathon'] = "London"
chicago['Marathon'] = "Chicago"
newyork['Marathon'] = "New York"
berlin['Marathon'] = "Berlin"
```

```python
lonchinyber = pd.concat([london, chicago, newyork, berlin], ignore_index=True)
```
