```python
def split_name_country(name):
    # Remove the "»" character
    name = name.replace('»', '').strip()

    # Split by comma to get last name and the rest
    parts = name.split(',')
    last_name = parts[0].strip() if len(parts) > 0 else ''

    # Check if there's a first name and country part
    if len(parts) > 1:
        first_name_country = parts[1].strip()

        # Extract country code by finding text in parentheses
        country = first_name_country[first_name_country.find("(")+1:first_name_country.find(")")]
        # Remove the country code and parentheses from the first name
        first_name = first_name_country.replace(f'({country})', '').strip()
    else:
        first_name, country = '', ''

    return last_name, first_name, country

# Apply the function to each DataFrame
for df in dfs:
    df[['Last Name', 'First Name', 'Country']] = df.apply(lambda row: split_name_country(row['Name']), axis=1, result_type='expand')
```
