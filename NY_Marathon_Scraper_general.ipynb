```python
import matplotlib.pyplot as plt
import pandas as pd
from bs4 import BeautifulSoup
import requests
import json
from IPython.display import HTML
import html
```

```python
soup = BeautifulSoup(raw_html2022, 'html.parser')
```

```python
pretty_soup = soup.prettify()
```

```python
name_nodes = soup.select('.name')
```

```python
def get_name(name_node):
  return name_node.get_text(strip=True)
```

```python
list_all_namedivs = []
for name_node in name_nodes:
  list_all_namedivs.append(get_name(name_node))
```

```python
index_to_remove = list_all_namedivs.index('TCS New York City Marathon 2021')
```

```python
list_all_namedivs = list_all_namedivs[index_to_remove + 1:]
```

```python
country_nodes = soup.select('[ng-if="eventFinisher.iaaf"]')
```

```python
def get_country(country_node):
  return country_node.get_text()
```

```python
gender_age_nodes = soup.select('[ng-if="eventFinisher.gender || eventFinisher.age"]')
```

```python
def get_gender(gender_age_node):
  return gender_age_node.get_text()[0]

def get_age(gender_age_node):
  return gender_age_node.get_text()[1:]
```

```python
place_nodes = soup.select('span.result.right-floated-item.mid-text span.num.ng-binding')
```

```python
def get_place(place_node):
  return place_node.get_text()
```

```python
time_nodes = soup.select('span.result.right-floated-item.long-text span.num.ng-binding')
```

```python
def get_time(time_node):
  return time_node.get_text()
```

```python
bib_nodes = soup.select('span.left-bordered.ng-scope[ng-if="eventFinisher.bib"] span.num.ng-binding')
```

```python
def get_bib(bib_node):
  return bib_node.get_text()
```

```python
list_all_countries = []
for country_node in country_nodes:
  list_all_countries.append(get_country(country_node))

list_all_gender = []
list_all_ages = []
for gender_age_node in gender_age_nodes:
  list_all_gender.append(get_gender(gender_age_node))
  list_all_ages.append(get_age(gender_age_node))

list_overall_places = []
for place_node in place_nodes:
  list_overall_places.append(get_place(place_node))

list_times = []
for time_node in time_nodes:
  list_times.append(get_time(time_node))

list_bibs = []
for bib_node in bib_nodes:
  list_bibs.append(get_bib(bib_node))
```

```python
assert len(list_all_namedivs) == len(list_all_gender) == len(list_all_ages) == len(list_all_countries) == len(list_bibs) == len(list_overall_places) == len(list_times)
```

```python
data = {
    'Name': list_all_namedivs,
    'Gender': list_all_gender,
    'Age': list_all_ages,
    'Country': list_all_countries,
    'BIB': list_bibs,
    'Overall Place': list_overall_places,
    'Time': list_times
}
```

```python
df = pd.DataFrame(data)
```
