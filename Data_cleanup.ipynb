```python
cmdf2000 = cmdf2000.drop(0)
```

```python
all_dfs = [(cmdf2000, 2000),
           (cmdf2001, 2001),
           (cmdf2002, 2002),
           (cmdf2003, 2003),
           (cmdf2004, 2004),
           (cmdf2005, 2005),
           (cmdf2006, 2006),
           (cmdf2007, 2007),
           (cmdf2008, 2008),
           (cmdf2009, 2009),
           (cmdf2010, 2010),
           (cmdf2011, 2011),
           (cmdf2012, 2012),
           (cmdf2013, 2013),
           (cmdf2014, 2014),
           (cmdf2015, 2015),
           (cmdf2016, 2016),
           (cmdf2017, 2017),
           (cmdf2018, 2018),
           (cmdf2019, 2019),
           (cmdf2021, 2021),
           (cmdf2022, 2022),
           (cmdf2023, 2023)]
```

```python
for df in all_dfs:
    # Extract 'Gender' and 'Age' from the 'Last Name, First Name' column, making age optional
    df[['Gender', 'Age']] = df['Last Name, First Name'].str.extract(r'\(([MF])(\d*)\)')

    # Replace any extracted age that is an empty string with a default value or NaN
    # You can define default_age to be whatever makes sense for your context, e.g., NaN, a specific integer, etc.
    default_age = None  # or np.nan, or a default integer
    df['Age'] = df['Age'].replace('', default_age)

    # Remove the parentheses and the gender-age information from the 'Last Name, First Name' column
    df['Last Name, First Name'] = df['Last Name, First Name'].str.replace(r'\s*\([MF]\d*\)', '', regex=True)
```
