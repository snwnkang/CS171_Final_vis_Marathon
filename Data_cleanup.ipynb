```python
#create DIV
age_brackets = {
    'M20-24': (20, 24),
    'M25-29': (25, 29),
    'M30-34': (30, 34),
    'M35-39': (35, 39),
    'M40-44': (40, 44),
    'M45-49': (45, 49),
    'M50-54': (50, 54),
    'M55-59': (55, 59),
    'F20-24': (20, 24),
    'F25-29': (25, 29),
    'F30-34': (30, 34),
    'F35-39': (35, 39),
    'F40-44': (40, 44),
    'F45-49': (45, 49),
    'F50-54': (50, 54),
    'F55-59': (55, 59),
}
```

```python
def div_assign(age, gender):
  for div, (start_age, end_age) in age_brackets.items():
    if start_age <= age <= end_age and div.startswith(gender):
      return div
  return 'Unknown'
```

```python
dfs_missing_div = [cmdf2005, cmdf2007]

for df in dfs_missing_div:
  df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
  df['DIV'] = df.apply(lambda row: div_assign(row['Age'], row['Gender']), axis=1)
```
