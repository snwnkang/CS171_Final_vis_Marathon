```python
import matplotlib.pyplot as plt
import pandas as pd
from bs4 import BeautifulSoup
import requests
import json
from IPython.display import HTML
import html
```

```python
chicago2014 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2014.csv")
chicago2015 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2015.csv")
chicago2016 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2016.csv")
chicago2017 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2017.csv")
chicago2018 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2018.csv")
chicago2019 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2019.csv")
chicago2021 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2021.csv")
chicago2022 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2022.csv")
chicago2023 = pd.read_csv("https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/chicago_data/Chicago_Marathon2023.csv")
```

```python
chicago2022
```

```python
combined_chicago = pd.concat(
    [chicago2014, chicago2015, chicago2016, chicago2017, chicago2018, chicago2019, chicago2021, chicago2022, chicago2023],
    ignore_index=True
)
```

```python
combined_chicago
```

```python
combined_chicago.to_csv("combined_Chicago_data.csv", index=False)
```
