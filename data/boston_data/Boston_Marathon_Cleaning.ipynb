```python
# Create list of relevant years in the dataframe
years = list(range(2014, 2020))

# Use 2000 to begin the dataframe
df = pd.DataFrame(columns=pd.read_csv('results2014.csv').columns)

# Iterate over years to combine all dataframes
for year in years:
    
    # Handle case where results were interrupted by 2013 terrorist attack
    if year == 2013:
        filename = 'results2013_without-diverted.csv'
    else:
        filename = f'results{year}.csv'
    
    tdf = pd.read_csv(filename, error_bad_lines=False)
    print(tdf)
    tdf['year'] = year
    
    df = pd.concat([df, tdf], axis=0)

older_df = df
```

```python
elite_bib = 200

# Create list of relevant years in the dataframe
years = list(range(2015, 2020))

# Use 2000 to begin the dataframe
df = pd.DataFrame()

# Iterate over years to combine all dataframes
for year in years:
    print(year)
    
    # Handle case where results were interrupted by 2013 terrorist attack
    if year == 2013:
        filename = 'results2013_without-diverted.csv'
    else:
        filename = f'results{year}.csv'
    
    # print(filename)
    tdf = pd.read_csv(filename, error_bad_lines=False)
    print(tdf)
    tdf['year'] = year
    
    df = pd.concat([df, tdf], axis=0)
    
df['bib'] = pd.to_numeric(df['bib'], errors='coerce')
elite_df = df[df['bib'] < elite_bib]
elite_df.reset_index(drop=True, inplace=True)
print(elite_df)
```

```python
### Handle 2023 Data
df = pd.read_excel('Boston_2021_to_2023.xlsx', sheet_name='2023')
print(df)

# Extract name, gender, and age from combined string
pattern = r'(?P<Name>.+) \((?P<Gender>M|F)(?P<Age>\d+)\)'
df[['name', 'gender', 'age']] = df['Name_Sex_Age'].str.extract(pattern)
df.drop(columns=['Name_Sex_Age'], inplace=True)

# Extract city, state, and country from combined location string when applicable
def split_location(location):
    components = location.split(', ')
    
    if len(components) == 1:
        city, state, country = None, None, components[0]
    elif len(components) == 2:
        city, state, country = components[0], None, components[1]
    elif len(components) == 3:
        city, state, country = components[0], components[1], components[2]
    else:
        city, state, country = None, None, None
    
    return pd.Series([city, state, country], index=['city', 'state', 'country'])

df[['city', 'state', 'country']] = df['city_state_country'].apply(split_location)
df.drop(columns=['city_state_country'], inplace=True)

### Extract gender_result and division_result from combined string
def split_place(place):
    if isinstance(place, str):
        components = place.split('\xa0/\xa0')
        if len(components) == 1:
            gender_result, division_result = components[0], None
        elif len(components) == 2:
            gender_result, division_result = components[0], components[1]
        else:
            gender_result, division_result = None, None
    else:
        gender_result, division_result = place, None
    
    return pd.Series([gender_result, division_result], index=['gender_result', 'division_result'])


df[['gender_result', 'division_result']] = df['sex_place-div_place'].apply(split_place)
df.drop(columns=['sex_place-div_place'], inplace=True)

df['year'] = 2023

df_2023 = df
print(df_2023)
```

```python
### Handle 2022 Data
df = pd.read_excel('Boston_2021_to_2023.xlsx', sheet_name='2022')

### Extract name, gender, and age from combined string
pattern = r'(?P<Name>.+) \((?P<Gender>M|F)\)'
df[['name', 'gender']] = df['Name_Sex_Age'].str.extract(pattern)
df.drop(columns=['Name_Sex_Age'], inplace=True)

### Extract gender_result and division_result from combined string
df[['gender_result', 'division_result']] = df['sex_place-div_place'].apply(split_place)
df.drop(columns=['sex_place-div_place'], inplace=True)

df['year'] = 2022

df_2022 = df
print(df_2022)
```

```python
### Handle 2021 Data
df = pd.read_excel('Boston_2021_to_2023.xlsx', sheet_name='2021')

### Extract name, gender, and age from combined string
pattern = r'(?P<Name>.+) \((?P<Gender>M|F)\)'
df[['name', 'gender']] = df['Name_Sex_Age'].str.extract(pattern)
df.drop(columns=['Name_Sex_Age'], inplace=True)

### Extract gender_result and division_result from combined string
df[['gender_result', 'division_result']] = df['sex_place-div_place'].apply(split_place)
df.drop(columns=['sex_place-div_place'], inplace=True)

df['year'] = 2021

df_2021 = df
print(df_2021)
```

```python
recent_df = pd.concat([df_2021, df_2022, df_2023], axis=0)

recent_df = recent_df.rename(columns={'name': 'display_name', 'country': 'country_citizenship'})
```

```python
older_df.dropna(how='all',axis=1,inplace=True)
print(recent_df.columns)
print(older_df.columns)
```

```python
def time_to_seconds(time_str):
    try:
        return pd.to_timedelta(time_str).total_seconds()
    except ValueError:
        return time_str.hour * 3600 + time_str.minute * 60 + time_str.second

recent_df['seconds'] = recent_df['official_time'].apply(time_to_seconds)
print(recent_df)
```

```python
# Combine recent and older dataframe
combo_df = pd.concat([recent_df,older_df],axis=0,ignore_index=True)

# Calculate approximate elite cutoffs for men
men_df = combo_df[combo_df['gender'] == 'M']
total_men = men_df['year'].value_counts()
total_men[2022] = 14318
total_men[2023] = 15171
men_cutoff = np.ceil(total_men*.001)

# Calculate approximate elite cutoffs for women
women_df = combo_df[combo_df['gender'] == 'F']
total_women = women_df['year'].value_counts()
total_women[2022] = 10590
total_women[2023] = 11405
women_cutoff = np.ceil(total_women*.001)

# Define years affected by covid so that it does not have a smaller pool
men_cutoff[2021] = 15
women_cutoff[2021] = 12

print(men_cutoff)
print(women_cutoff)
```

```python
# Convert seconds to numeric type so that it will work in this code
combo_df['seconds'] = pd.to_numeric(combo_df['seconds'], errors='coerce')

elite_df = pd.DataFrame()

for sex in ['M','F']:
    for year in combo_df['year'].unique():
        
        # Use specific cutoff series
        if sex == 'M':
            cutoff = men_cutoff
        else:
            cutoff = women_cutoff
        
        # Filter dataframe to get top "cutoff" runners for each classification
        filtered_df = combo_df[(combo_df['year'] == year) & (combo_df['gender'] == sex)]
        filtered_df = filtered_df.nsmallest(int(cutoff[year]), 'seconds')
        elite_df = pd.concat([elite_df,filtered_df],axis=0,ignore_index=True)
        
```

```python
elite_df.to_csv('Boston_elite_2014_to_2023.csv', index=False)
combo_df.to_csv('Boston_all_2014_to_2023.csv', index=False)
```
