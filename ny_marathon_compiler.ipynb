```python
import pandas as pd
```

```python
nydf2014 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2014.csv')
nydf2015 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2015.csv')
nydf2016 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2016.csv')
nydf2017 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2017.csv')
nydf2018 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2018.csv')
nydf2019 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2019.csv')
nydf2020 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2020.csv')
nydf2021 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2021.csv')
nydf2022 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2022.csv')
nydf2023 = pd.read_csv('https://raw.githubusercontent.com/snwnkang/CS171_Final_vis_Marathon/main/data/nymarathon_data/NY_Marathon2023.csv')
```

```python
#add a year column to all of them
nydf2014['Year'] = 2014
nydf2015['Year'] = 2015
nydf2016['Year'] = 2016
nydf2017['Year'] = 2017
nydf2018['Year'] = 2018
nydf2019['Year'] = 2019
nydf2020['Year'] = 2020
nydf2021['Year'] = 2021
nydf2022['Year'] = 2022
nydf2023['Year'] = 2023
```
